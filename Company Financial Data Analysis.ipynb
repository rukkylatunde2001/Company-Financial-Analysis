```python
# Import python libararies

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# Load csv data

df = pd.read_csv('Company_Financials_dataset.csv')
```

```python
# Display first few rows

df.head()
```

```python
#Check column types

df.dtypes
```

```python
# Display column names

df.columns
```

```python
## DATA CLEANING PROCESS
```

```python
# Remove whitespace from column names

df.columns.str.strip()
```

```python
# signs and symbols from columns

df = df.replace({'\$':'',',':'','-':''}, regex = True)
```

```python
df.head()
```

```python
# convert date column to datetime

df['Date'] = pd.to_datetime(df['Date'])     

df.head()
```

```python
# Define columns to clean. Replace empty strings with NaN and convert data type

cols = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts', 'Sales', 'COGS', 'Profit']

# Replace empty strings or whitespace with NaN
df[cols] = df[cols].replace('', float('nan')).apply(pd.to_numeric, errors='coerce')

# Convert specified columns to float
df[cols] = df[cols].astype(float)
```

```python
# Check data types

df.dtypes
```

```python
# checking for missing values

df.isna().sum()
```

```python
# checking null values

df[df['Profit'].isnull()]
```

```python
# Replace NaN with 0 for 'Discounts' and 'Profit' columns

df[['Discounts', 'Profit']] = df[['Discounts', 'Profit']].fillna(0)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
# Analysis

# Total sales & profit by segment
# Total sales & profit by country
# compare sales and profit across market segments
# Analyze trends within each market segment
```

```python
# Total sales by segment

Total_sales_by_segment = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
Total_sales_by_segment
```

```python
# Total profits by segment

Total_profits_by_segment = df.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
Total_profits_by_segment
```

```python
# Total profits by country

Total_profits_by_country = df.groupby('Country')['Profit'].sum().sort_values(ascending=False)
Total_profits_by_country
```

```python
# Total sales by country
Total_sales_by_country = df.groupby('Country')[['Sales', 'Profit']].sum().sort_values(by=['Sales', 'Profit'], ascending=False)
Total_sales_by_country
```

```python
plt.figure(figsize=(11, 6))

plt.subplot(1, 2, 1)
sns.barplot(x='Segment', y='Sales', data=df.groupby('Segment')['Sales'].mean().reset_index())
plt.xlabel('Market Segment')
plt.ylabel('Average Sales')
plt.title('Average Sales by Market Segment')
plt.xticks(rotation=45)


plt.subplot(1, 2, 2)
sns.barplot(x='Segment', y='Profit', data=df.groupby('Segment')['Profit'].mean().reset_index())
plt.xlabel('Market Segment')
plt.ylabel('Average Profit')
plt.title('Average Profit by Market Segment')
plt.xticks(rotation=45)
```

```python
# The most profitable segment with the higest avearge sales is the small 
business segment.
```

```python
# Sales trend

plt.figure(figsize=(11, 6))
sns.lineplot(x ='Date', y ='Sales', hue = 'Segment', data = df)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales over time by Market Segment')
plt.xticks(rotation=45)
```

```python
# Profits trend

plt.figure(figsize=(14, 7))
sns.lineplot(x=df['Month Name'], y='Profit', hue='Segment', data=df)
plt.xlabel('Month')
plt.ylabel('Profits')
plt.title('Profits over time by Market Segment')
plt.xticks(rotation=45, ha='right')  # Adjust rotation and alignment for better visibility
```

# Findings

The highest financial contributor is the governemnt segment which has the highest total sales and profit. This mean the government segment contribute the most to the overall revenue and profit of the company.

The most efficient and profitable segment is the small business segment which on the other hand  has the highest avearge sales and profit. We could say this segment yields better returns. 

The United States generated the highest sales while France has the highest profit.

The line plot shows the trends in profit. The government and small business are on top of the chart with a high spike in small business and drop in government in February.
